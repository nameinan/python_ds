```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
plt.plot(x,x**2)
plt.plot(x,x**3)
plt.plot(x,x*2)
plt.plot(x,2**x)
plt.show()
```
