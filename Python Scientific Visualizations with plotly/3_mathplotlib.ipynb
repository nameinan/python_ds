```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(5)
x
```

```python
y=x
```

```python
y
```

```python
plt.plot(x,y,'o--')
plt.plot(x,-y,'o-')
plt.title('y=x and y=-x')
plt.show()
```

```python
N=10
x=np.linspace(0,10,N)
y=x
```

```python
plt.plot(x,y,'o--')
plt.axis('on')

plt.title('y=x')
plt.show()
```

```python
y=np.logspace(0,1,N)
y
plt.plot(x,y,'o--')
```

```python
y=np.geomspace(0.1,100,N)
y
plt.plot(x,y,'o--')
```
