```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
#plt.grid(False)
print(plt.axis())
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
plt.axis([0,2,0,8])
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8])
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.xlabel('x= np.arrange(3)')
plt.xlabel('y= f(x)')
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8])
plt.title('Simple plot demonstration')
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,label='x**2')
plt.plot(x,x**3,label='x**3')
plt.plot(x,2*x,label='2*x')
plt.plot(x,2**x,label='2**x')
plt.legend()
plt.xlabel('x= np.arrange(3)')
plt.xlabel('y= f(x)')
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8])
plt.title('Simple plot demonstration')
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.legend(['x**2','x**3','2*x','2**x'])
plt.xlabel('x= np.arrange(3)')
plt.xlabel('y= f(x)')
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8])
plt.title('Simple plot demonstration')
plt.show()
```

```python
x= np.arange(3)
plt.plot(x,x**2,x,x**3,x,2*x,x,2**x)
plt.legend(['x**2','x**3','2*x','2**x'],loc='upper right')
plt.xlabel('x= np.arrange(3)')
plt.xlabel('y= f(x)')
plt.grid(True)
plt.xlim([0,2])
plt.ylim([0,8])
plt.title('Simple plot demonstration')
plt.savefig('test.png')
plt.show()
```
